```python
def extract_year_month(date_str):
    date_str_simplified = date_str[:24]  # Removing the time zone information
    date_obj = datetime.strptime(date_str_simplified, "%a %b %d %Y %H:%M:%S")
    return date_obj.year, date_obj.month
```

```python
year_month = df_2['saledate'].apply(extract_year_month)
```
